```python
import pandas as pd
df = pd.read_csv('/Users/maxsonntag/Documents/GitHub/SOC_predictor/data/lucas_db/LUCAS-SOIL-2018.csv')
df
```

```python
df = df.loc[df['LU1_Desc'] == 'Agriculture (excluding fallow land and kitchen gardens)']
df
```

```python
df['OC'].value_counts()
```

```python
# Convert the 'OC' column to numeric, setting errors='coerce' to handle non-numeric values as NaN
df['OC'] = pd.to_numeric(df['OC'], errors='coerce')

# Now filter the DataFrame for values where 'OC' is greater than 1000
filtered_df = df[df['OC'] < 150]
filtered_df
```

```python
import pandas as pd

# Load the feature table
feature_table = pd.read_csv('/Users/maxsonntag/Documents/GitHub/SOC_predictor/data/feature_table.csv')
# Round the latitude and longitude columns in the original table to 5 decimals
df['TH_LAT_rounded'] = df['TH_LAT'].round(5)
df['TH_LONG_rounded'] = df['TH_LONG'].round(5)

# Convert OC from g/kg to %
df['OC'] = pd.to_numeric(df['OC'], errors='coerce')
df['soc_percent'] = df['OC'] / 10

# Merge feature table and original data based on rounded lat/long
merged_data = feature_table.merge(
    df[['TH_LAT_rounded', 'TH_LONG_rounded', 'soc_percent']],
    left_on=['lat', 'long'],
    right_on=['TH_LAT_rounded', 'TH_LONG_rounded'],
    how='left'
)

# Replace soc_in_percent with OC_percent from original data
merged_data['soc_in_percent'] = merged_data['soc_percent']
```

```python
import matplotlib.pyplot as plt

# Replace 'column_name' with your column and 'df' with your DataFrame
plt.figure(figsize=(10, 6))
plt.hist(merged_data['soc_percent'], bins=30, edgecolor='k', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Column Values')
plt.show()
```

```python
# Drop the 'mean_temp_2017' column
#merged_data = merged_data.drop(columns=['mean_temp_2017'])

merged_data = merged_data.drop(columns=['soc_percent'])
merged_data = merged_data.drop(columns=['TH_LAT_rounded'])
merged_data = merged_data.drop(columns=['TH_LONG_rounded'])

# Save the merged data to a CSV file
merged_data.to_csv('/Users/maxsonntag/Documents/GitHub/SOC_predictor/data/feature_table_fixed.csv', index=False)
```
