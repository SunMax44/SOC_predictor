```python
import pandas as pd

# Load the CSV with the semicolon delimiter
df = pd.read_csv('/Users/maxsonntag/Documents/GitHub/SOC_predictor/data/bze/site.csv', delimiter=';')

# Replace all commas with dots in all columns
df = df.applymap(lambda x: str(x).replace(',', '.') if pd.notnull(x) else x)

# Try to convert columns back to numeric types where possible
df = df.apply(pd.to_numeric, errors='ignore')

# Save the cleaned DataFrame to a new CSV file
df.to_csv('/Users/maxsonntag/Documents/GitHub/SOC_predictor/data/bze/site.csv', 
          index=False, 
          sep=';')
```
